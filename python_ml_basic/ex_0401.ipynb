## 선형회귀 분석

미국 보스톤 부동산 시세 데이터 (sklearn 패키지 제공).

#### 필요한 패키지를 불러온다:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
```

#### 데이터를 불러온다:

```python
data = load_boston()
```

```python
data.keys()
```

```python
print(data['DESCR'])
```

```python
# 입력 데이터
X = data['data']
X[0] # 13 항목의 데이터들이 들어옴. Xi값 i = 0~13
```

```python
# 컬럼 명들 ( 독립변수명들 )
header = data['feature_names']
```

```python
# 결과 데이터
Y = data['target']
Y
```

```python
Y = Y.reshape(-1, 1) # 행 벡터인 array Y를 열 벡터로 바꿔준다.
```

```python
header
```

#### 데이터 프레임을 만들어 본다:

```python
df = pd.DataFrame(np.append(X, Y, axis = 1))
df
```

```python

df.columns = np.append(header,'PRICE')
df
```

```python
df.head(5)
```

```python
df.tail(5)
```

#### 여러 통계치를 계산해 본다:

```python
df.describe()
```

```python
np.round(df.corr(),3) # 상관계수 구함, 3자리 까지
```

#### 선형회귀 적용:

```python
# lm ?? 
lm = LinearRegression()
```

```python
lm.fit(X, Y)  # X : 입력, Y : 결과
# fit 함수는 모델을 학습시켜 모델 객체에 저장한다.
# LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)

# When fit_intercept=False, the intercept is forced to the origin (0, 0).
```

```python
lm.coef_
# 계수 (베타)
```

```python
pd.DataFrame(lm.coef_, columns = header)
# 각 컬럼(독립변수:Xi)별 계수 (베타)를 보여줌
```

```python
lm.intercept_
```

```python
plt.scatter(X[:,5],Y[:,0],c = 'g',s=15,alpha=0.5)
plt.xlabel('RM')
plt.ylabel('PRICE')
plt.show()
```

#### 예측 가격 (in-sample testing):

```python
Y_pred = lm.predict(X)
```

```python
plt.scatter(Y[:,0],Y_pred,c = 'blue', s=15, alpha=0.5)
plt.xlabel('REAL PRICE')
plt.ylabel('PREDICTED PRICE')
plt.show()
```

#### 결정계수 ($R^2$):

```python
lm.score(X,Y)
```

#### 학습과 시험 (in and out-of-sample testing):

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=5) # random_state = random.seed ㅎㅎ 그래서 5라는 값은 의미있는 값은 아님
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
lm = LinearRegression()
lm.fit(X_train,Y_train)
Y_pred_train = lm.predict(X_train)
Y_pred_test = lm.predict(X_test)
```

```python
print('Training MSE is:' + str(np.mean((Y_train - Y_pred_train)**2)))
print('Testing MSE is:' + str(np.mean((Y_test - Y_pred_test)**2)))
```

#### 잔차:

```python
# 잔차 분석
plt.scatter(np.arange(Y_train.size),Y_pred_train-Y_train,c = 'red', s=15, alpha=0.5)
plt.xlabel('N#')
plt.ylabel('Train Residual')
plt.show()
# 200부근에 음수값이 하나 튀는구나!
```

```python
plt.scatter(np.arange(Y_test.size),Y_pred_test-Y_test,c = 'red', s=15, alpha=0.5)
plt.xlabel('N#')
plt.ylabel('Test Residual')
plt.show()
```
