### 시계열 예측

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from statsmodels.tsa.holtwinters import ExponentialSmoothing
%matplotlib inline
```

```python
os.chdir(r"C:\Users\admin\Desktop\20180709_기본수학과 이론\data")
```

```python
df = pd.read_csv('data_time_series.csv', header='infer',encoding = 'latin1')
```

```python
df.plot()
plt.show()
```

```python
data = df['Value']
model = ExponentialSmoothing(data.iloc[0:35], seasonal='mul',seasonal_periods=6).fit()
pred = model.predict(start= 36, end=47)
plt.plot(data,color='blue')
plt.plot(pred,color='red')
plt.show()
```
